# **Data Cleaning & Encoding**

```python
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
```

```python
df = pd.get_dummies(df, columns=['Blood_Type', 'marital_Status'], drop_first=True)
```

```python
df = df.drop(columns=['Nationality'])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Disease_Class_Encoded'] = le.fit_transform(df['Disease_Class'])


for i, label in enumerate(le.classes_):
    print(f"{label} -> {i}")
```

# **Define Features (X)**
We exclude the raw target columns. `Patient_Severity_Score` is excluded here because it becomes its own regression target later.

```python
columns_to_exclude = ['Disease_Class', 'Disease_Class_Encoded',  'Patient_Severity_Score']
X = df.drop(columns=columns_to_exclude)
print("Number of Features:", X.shape[1])
print(X.columns.tolist())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_sample_weight
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, mean_absolute_error
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestRegressor
import joblib
```

# **Model 1: Healthy vs Has_Disease (Binary Classification)**
First stage: decide whether the patient shows signs of any condition at all (`Disease_Class != 'Other'`) before identifying which specific disease it might be.


```python
y_binary = (df['Disease_Class'] != 'Other').astype(int)

X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(
    X, y_binary, test_size=0.2, random_state=42, stratify=y_binary
)

sample_weights_b = compute_sample_weight(class_weight='balanced', y=y_train_b)

model_binary = XGBClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=6,
    random_state=42,
    eval_metric='logloss'
)

model_binary.fit(X_train_b, y_train_b, sample_weight=sample_weights_b)

y_pred_b = model_binary.predict(X_test_b)
print(classification_report(y_test_b, y_pred_b, target_names=['Other', 'Has_Disease']))
```

# **Model 2: Disease Type (Diabetes / Heart Disease / Hypertension)**
Second stage: trained only on patients who are NOT 'Other', so it always commits to one of the 3 diseases. Should only be called on patients Model 1 already flagged as likely-disease.

```python
mask_no_other = df['Disease_Class'] != 'Other'

X_no_other = X[mask_no_other]
y_no_other = df.loc[mask_no_other, 'Disease_Class_Encoded']

print("Number of rows after deleting Other:", X_no_other.shape[0])
print(y_no_other.value_counts())

X_train_no, X_test_no, y_train_no, y_test_no = train_test_split(
    X_no_other, y_no_other, test_size=0.2, random_state=42, stratify=y_no_other
)

sample_weights_no = compute_sample_weight(class_weight='balanced', y=y_train_no)

model_no_other = XGBClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=6,
    random_state=42,
    eval_metric='mlogloss'
)

model_no_other.fit(X_train_no, y_train_no, sample_weight=sample_weights_no)

y_pred_no = model_no_other.predict(X_test_no)

from sklearn.preprocessing import LabelEncoder
le_no_other = LabelEncoder()
le_no_other.fit(df.loc[mask_no_other, 'Disease_Class'])

print(classification_report(y_test_no, y_pred_no, target_names=le_no_other.classes_))
```

# **Combined Prediction Pipeline**
Wraps Model 1 and Model 2 together so a single patient row goes through both stages automatically.

```python
def predict_patient(patient_features_df):
    """
    patient_features_df: single-row DataFrame with exactly the same columns as X.
    """
    has_disease = model_binary.predict(patient_features_df)[0]
    has_disease_proba = model_binary.predict_proba(patient_features_df)[0][1]

    if has_disease == 0:
        return {
            'classification': 'Healthy (Other)',
            'confidence': f"{(1-has_disease_proba)*100:.1f}%",
            'disease_type': '-'
        }
    else:
        disease_pred = model_no_other.predict(patient_features_df)[0]
        disease_proba = model_no_other.predict_proba(patient_features_df)[0]
        disease_name = le_no_other.inverse_transform([disease_pred])[0]
        confidence = disease_proba[disease_pred] * 100

        return {
            'classification': 'Needs checkup',
            'confidence': f"{has_disease_proba*100:.1f}%",
            'disease_type': disease_name,
            'disease_confidence': f"{confidence:.1f}%"
        }

# Sanity check on the first row of the test set
sample_patient = X_test_b.iloc[[0]]
result = predict_patient(sample_patient)
print("Prediction result:")
for key, value in result.items():
    print(f"{key}: {value}")

actual_idx = X_test_b.index[0]
print("\nActual ground truth:", df.loc[actual_idx, 'Disease_Class'])
```

# **Try the Pipeline on a Manually-Entered Patient**

```python

new_patient = X.iloc[[0]].copy()


new_patient['Age'] = 58
new_patient['Avg_SystolicBP'] = 165      
new_patient['Max_SystolicBP'] = 180
new_patient['Avg_DiastolicBP'] = 100
new_patient['Avg_HeartRate'] = 110       
new_patient['Max_HeartRate'] = 125
new_patient['Avg_MAP'] = 121
new_patient['Avg_Glucose'] = 90          
new_patient['Avg_Temperature'] = 37.2
new_patient['Max_Temperature'] = 37.5
new_patient['Avg_O2Sat'] = 94
new_patient['BMI'] = 31
new_patient['Chronic_Disease_Count'] = 2

result = predict_patient(new_patient)
print("Predicition Result")
for key, value in result.items():
    print(f"{key}: {value}")
```

# **Model 3: Patient Severity Score (Regression)**
A separate regression model that estimates the continuous `Patient_Severity_Score` (0-100) directly from the patient's features, independent of the disease classification above.

```python
y_severity = df['Patient_Severity_Score']

X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(
    X, y_severity, test_size=0.2, random_state=42
)

model_severity = RandomForestRegressor(n_estimators=150, random_state=42, n_jobs=-1)
model_severity.fit(X_train_s, y_train_s)

y_pred_s = model_severity.predict(X_test_s)
mae = mean_absolute_error(y_test_s, y_pred_s)
print(f"Mean Absolute Error for Severity Score prediction: {mae:.2f} points (out of 100)")
```
