## **Missing Valuse**

```python
df.isnull().sum()
```

```python
df['Chronic_Disease_Count']=df['Chronic_Disease_Count'].fillna(0)
```

```python
df['Chronic_Disease_Count'].isnull().sum()
```

```python
df['Avg_Glucose']=df['Avg_Glucose'].fillna(df['Avg_Glucose'].median())
df['Max_Glucose']=df['Max_Glucose'].fillna(df['Max_Glucose'].median())
```

```python
df['Avg_Glucose'].isnull().sum()
df['Max_Glucose'].isnull().sum()
```

```python
#Overweight and obese patients show a higher frequency of extreme glucose values, 
# indicating a potential association between elevated BMI and diabetes risk.
plt.figure(figsize=(5,5))
sns.boxplot(x='BMI_Category', y='Avg_Glucose', data=df)
plt.xticks(rotation=45)
plt.title('BMI Category vs Average Glucose')
plt.show()
```

```python
#Older patients tend to exhibit multiple chronic diseases more frequently compared to younger patients.
plt.figure(figsize=(5,5))
sns.scatterplot(x='Age', y='Chronic_Disease_Count', data=df)
plt.title('Age vs Chronic Disease Count')
plt.show()
```

```python
#Elderly patients exhibit more extreme glucose levels, suggesting a higher risk of diabetes among older age groups.
sns.boxplot(x='Age_Group', y='Avg_Glucose', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
#Patients with high glucose levels tend to have higher BMI values,
# indicating a relationship between obesity and glucose abnormalities.
sns.boxplot(x='Glucose_Flag', y='BMI', data=df)
plt.show()
```

```python
#Patients with high cholesterol tend to have elevated systolic blood pressure, 
# suggesting increased cardiovascular risk.
sns.boxplot(x='HighCholesterol_Flag', y='Avg_SystolicBP', data=df)
plt.show()
```

```python
#Patients with multiple chronic diseases tend to experience longer hospital stays.
sns.scatterplot(x='Chronic_Disease_Count', y='Avg_Stay_Days', data=df)
plt.show()
```

```python
#Systolic blood pressure tends to increase with age, particularly among older patients.
sns.scatterplot(x='Age', y='Avg_SystolicBP', data=df)
plt.show()
```

```python
#Higher BMI values are associated with elevated systolic blood pressure levels.
sns.scatterplot(x='BMI', y='Avg_SystolicBP', data=df)
plt.show()
```
